```python
import pandas as pd
import numpy as np
from scipy import stats



```

```python
data= [100, 20, 5, 30, 45, -100, 45]
the_average=np.mean(data)
print("the average= ", the_average)

the_median= np.median(data)
print("the median= ", the_median)# skew and non skew values

the_mode= stats.mode(data)
print("the mode= ", the_mode)
```

```python
#variance and standard deviation
variance_= np.var(data)
pstd_= np.std(data)
print(f"the variance= ", {variance_} ," and the standard deviation= ", {pstd_})
```
