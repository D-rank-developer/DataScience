```python
import pandas as pd     
data=pd.DataFrame({'Name': ["Anna", "Karol","John","Alice","Kevin","Sanna", "Bob", "Alice"],
                   'Age': [35,30,57,65,25,19,20,65],
                   'Salary': [20310, 60003, 145160, 170000,30000,10310,220003,1200010],
                    'Departments':["Tech", "Tech","Tech", "Healthcare", "Operations","Operations", "Tech" ,"Tech"]})
```

```python
#sorting values and the  order according to descending order
data_sort_values=data.sort_values(by="Salary", ascending= True)
print(data_sort_values)

```

```python
data_sort_values2=data.sort_values(by="Salary", ascending= False)
print(data_sort_values2)
```

```python
#grouping data using the pandas function groupby--done by name and others
print(data.groupby("Departments").count())
print(data.groupby("Departments")["Name"].count())#exact column of the grouping
print(data.groupby("Departments")["Salary"].mean())#Calculating average salaries
print(data.groupby("Departments")["Salary"].min())
print(data.groupby("Departments")["Salary"].max())
print(data.groupby("Departments")["Age"].mean())
```

```python
#Filtering in python using the [] and specify the condition and specify the threshold
#we can still add more conditions (multiple filtering)
print(data[data["Salary"]>100000])

print(data[(data["Salary"]>100000) & (data["Salary"]<200000)]) #two conditions
```

```python
#use of isin
print(data[data["Age"].isin([20,45])])
```
